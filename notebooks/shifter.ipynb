```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
BOUNDARY_DIR = "../data/boundary/"
```

```python
def graph0(a, b):
    fig, ax = plt.subplots(figsize=(12, 10))
    plt.xlabel('$\it{x}$ [px]', fontsize=28)
    plt.ylabel('$\it{y}$ [px]', fontsize=28)
    plt.xticks(fontsize=24)
    plt.yticks(fontsize=24)
    plt.xlim(0, 1280)
    plt.ylim(0, 1024)
    plt.grid()

    plt.scatter(b[1:], 1024-a[1:], c='k')

    plt.show()

def graph1(a, b, x0, y0):
    fig, ax = plt.subplots(figsize=(12, 10))
    plt.xlabel('$\it{x}$ [px]', fontsize=28)
    plt.ylabel('$\it{y}$ [px]', fontsize=28)
    plt.xticks(fontsize=24)
    plt.yticks(fontsize=24)
    plt.xlim(-800, 800)
    plt.ylim(-50, 1024)
    plt.grid()

    plt.hlines(0, -800, 800, colors='r', linewidth=3)
    plt.vlines(0, -50, 1024, colors='r', linewidth=3)

    plt.scatter(b[1:]+x0, 1024-a[1:]+y0, c='b')

    plt.show()
```

- 175, 215, 18

```python
data = np.loadtxt(BOUNDARY_DIR + "u_175_q_215_d_18.csv", delimiter=',')
A, B = data[0], data[1]
graph0(A, B)
```

```python
graph1(A, B, -880, -120)
```
