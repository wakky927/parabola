```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
BOUNDARY_DIR = "../data/boundary/"
SHIFT_DIR = "../data/shift/"

u_list = [175, 200, 225, 250]
q_list = [215, 304, 429]
d_list = [18, 24, 30]
```

```python
def graph0(a, b):
    fig, ax = plt.subplots(figsize=(12, 10))
    plt.xlabel('$\it{x}$ [px]', fontsize=28)
    plt.ylabel('$\it{y}$ [px]', fontsize=28)
    plt.xticks(fontsize=24)
    plt.yticks(fontsize=24)
    plt.xlim(0, 1280)
    plt.ylim(0, 1024)
    plt.grid()

    plt.scatter(b[1:], 1024-a[1:], c='k')

    plt.show()

def graph1(a, b, x0, y0):
    fig, ax = plt.subplots(figsize=(12, 10))
    plt.xlabel('$\it{x}$ [px]', fontsize=28)
    plt.ylabel('$\it{y}$ [px]', fontsize=28)
    plt.xticks(fontsize=24)
    plt.yticks(fontsize=24)
    plt.xlim(-800, 800)
    plt.ylim(-50, 1024)
    plt.grid()

    plt.hlines(0, -800, 800, colors='r', linewidth=3)
    plt.vlines(0, -50, 1024, colors='r', linewidth=3)

    plt.scatter(b[1:]+x0, 1024-a[1:]+y0, c='b')

    plt.show()

def graph2(a, b):
    fig, ax = plt.subplots(figsize=(12, 10))
    plt.xlabel('$\it{x}$ [px]', fontsize=28)
    plt.ylabel('$\it{y}$ [px]', fontsize=28)
    plt.xticks(fontsize=24)
    plt.yticks(fontsize=24)
    plt.xlim(-800, 800)
    plt.ylim(-50, 1024)
    plt.grid()

    plt.hlines(0, -800, 800, colors='r', linewidth=3)
    plt.vlines(0, -50, 1024, colors='r', linewidth=3)

    plt.scatter(b, a, c='g')

    plt.show()
```

```python
def shifter(i, a, b):
    x0_y0 = np.loadtxt(BOUNDARY_DIR + "x0_y0.csv", delimiter=',')
    x0, y0 = x0_y0[i, 0], x0_y0[i, 1]

    return 1024-a+y0, b+x0
```
