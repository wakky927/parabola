```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
freq_cnt_file = "../data/result/freq_cnt_u_175_q_215_d_18_ppm_0.csv"
freq_cnt_file2 = "../data/result/2_u_175_q_215_d_18_ppm_0.csv"
img_file = "../data/result/00000000.csv"
data = np.loadtxt(freq_cnt_file, delimiter=',')
data2 = np.loadtxt(freq_cnt_file2, delimiter=',')
img = np.loadtxt(img_file, delimiter=',')
```

```python
X, Y = np.mgrid[-900:380, -824:200]

length = 10
x1 = 505
y1 = 447
x2 = 506
y2 = 530

z = 10 / np.sqrt((x1 - x2)**2 + (y1 - y2)**2)

m = 3142
theta_original = np.linspace(0, 2*np.pi, 1001)
theta = np.append(theta_original[1:500], theta_original[501:-1])

r = m * (np.pi - theta) / 175 / np.sin(theta)
```

```python
fig, ax = plt.subplots(figsize=(12, 10))
ax.xaxis.label.set_color('w')
ax.yaxis.label.set_color('w')
ax.tick_params(axis='x', colors='w')
ax.tick_params(axis='y', colors='w')
plt.xlabel('$\it{x}$ [px]', fontsize=28)
plt.ylabel('$\it{y}$ [px]', fontsize=28)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.xlim(-900, 380)
plt.ylim(-200, 824)

plt.pcolormesh(X, -Y, data.T, cmap='jet', alpha=1)
pp = plt.colorbar()
plt.clim(0, 1.0)

# plt.plot(r * np.sin(theta)/z, (r * np.cos(theta) + m / 175)/z, color='r', linewidth=3)
```

```python
fig, ax = plt.subplots(figsize=(12, 10))
ax.xaxis.label.set_color('w')
ax.yaxis.label.set_color('w')
ax.tick_params(axis='x', colors='w')
ax.tick_params(axis='y', colors='w')
plt.xlabel('$\it{x}$ [px]', fontsize=28)
plt.ylabel('$\it{y}$ [px]', fontsize=28)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.xlim(-900, 380)
plt.ylim(-200, 824)

plt.pcolormesh(X, -Y, img.T / 255, cmap='gray', alpha=1)
plt.plot(r * np.sin(theta)/z, (r * np.cos(theta) + m / 175)/z, color='r', linewidth=3)
```
